```python
import geopandas as gpd
import pandas as pd
```

```python
# データを変数に入れる
akita_ku = gpd.read_file('N03-23_05_230101.geojson')
# とりあえずマップとしてアウトプット
akita_ku.plot()
```

```python
# 最初の５列は？
akita_ku.head()
```

```python
# データの情報
akita_ku.info()
```

```python
akita_pop = pd.read_csv('akita.csv')
```

```python
akita_pop.head()
```

```python
akita_pop.info()
```

```python
# rename
akita_ku.rename(columns={'N03_007':'KEY_CODE'},inplace=True)
```

```python
# check the new fieldname
akita_ku
```

```python
# OBJECTからintに変える
akita_ku['KEY_CODE'] = akita_ku['KEY_CODE'].astype(int)
# checkする
akita_ku.info
```

```python
# ここでようやく merge コマンドで統合させて、新しい変数に入れる
akita_ku_pop = akita_ku.merge(akita_pop,on='KEY_CODE')
```

```python
# check!
akita_ku_pop
```

```python
akita_ku_pop.info()
```

```python
# change from OBJECT to int
akita_ku_pop['人口総数'] = akita_ku_pop['人口総数'].astype(int)
```

```python
# check
akita_ku_pop.info()
```

```python
akita_ku_pop.plot(column='人口総数',legend=True)
```
